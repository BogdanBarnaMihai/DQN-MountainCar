```python
import gymnasium as gym 
import math,random,numpy as np
from collections import deque,namedtuple
import torch,torch.nn as nn,torch.optim as optim 
import torch.nn.functional as f 

device=torch.device("cuda" if torch.cuda.is_available() else "cpu")

class MoutainCarRewardWrapper(gym.RewardWrapper):
    def reward(self,reward):
        pos,vel=self.env.unwrapped.state 
        if pos>0.45: reward+=100
        elif pos>=0.3: reward+=10
        elif pos>0.1:reward+=3
        reward+=abs(vel)*5
        return reward

transition=namedtuple("transition" , ("state","action","next_state","reward")) 
    
class Replaymemory():
    def __init__(self,cap): self.mem=deque([],maxlen=cap)
    def push(self,*a): self.mem.append(transition(*a))
    def sample(self,n) : return random.sample(self.mem,n)
    def __len__(self):return len(self.mem)

class DQN(nn.Module):
    def __init__(self,n_obs,n_act):
        super().__init__() 
        self.net=nn.Sequential(
            nn.Linear(n_obs,128),nn.ReLU(),
            nn.Linear(128,128),nn.ReLU(),
            nn.Linear(128,n_act)
        )
    def forward(self,x):return self.net(x)


lr=5e-4
gamma=0.99
learning_starts=1_000
eps_end=0.05
eps_start=1.0
eps_fraction=0.5
batch_size=128
buffer_size=100_000
total_steps=500_000
target_update=1_000
train_freq=4

env=MoutainCarRewardWrapper(gym.make("MountainCar-v0"))
n_obs=env.observation_space.shape[0]
n_act=env.action_space.n 

policy_net=DQN(n_obs,n_act).to(device)
target_net=DQN(n_obs,n_act).to(device)

target_net.load_state_dict(policy_net.state_dict())
target_net.eval() 

optimizer=optim.Adam(policy_net.parameters(),lr=lr)
memory=Replaymemory(buffer_size)

def get_eps(step):
    progress=step/(total_steps*eps_fraction)
    return max(eps_end,eps_start-(eps_start-eps_end)*progress)

def select_action(state,step):
    if random.random()<get_eps(step):
        return env.action_space.sample() 
    with torch.no_grad(): 
        return policy_net(state).argmax().item() 
    
def optimize():
    if len(memory)<batch_size: return 
    batch=transition(*zip(*memory.sample(batch_size)))
    non_final_mask=torch.tensor([s is not None for s in batch.next_state],device=device,dtype=torch.bool)
    non_final_next=torch.cat([s for s in batch.next_state if s is not None])
    nxt=torch.zeros(batch_size,device=device)
    ab=torch.cat(batch.action)
    sb=torch.cat(batch.state)
    rb=torch.cat(batch.reward)
    q=policy_net(sb).gather(1,ab)
    with torch.no_grad():
        nxt[non_final_mask]=target_net(non_final_next).max(1).values
    loss=f.smooth_l1_loss(q,(rb+gamma*nxt).unsqueeze(1))
    optimizer.zero_grad()
    loss.backward()
    torch.nn.utils.clip_grad_value_(policy_net.parameters(),100)
    optimizer.step() 

state,_=env.reset() 
st=torch.tensor(state,dtype=torch.float32,device=device).unsqueeze(0)

ep=0
ep_start=0
hundredreward=0.0

for step in range(1,total_steps+1):
    action=select_action(st,step)
    obs,reward,terminated,truncated,_=env.step(action)
    hundredreward+=reward
    nst=None if terminated else\
        torch.tensor(obs,dtype=torch.float32,device=device).unsqueeze(0)
    memory.push(st,torch.tensor([[action]],dtype=torch.long,device=device),nst,torch.tensor([reward]))    
    state,st=obs,nst
    done=terminated or truncated
    if done: 
        ep+=1
        if ep%100==0:
            print(f"episodes from {ep-100} to {ep} have a mean reward of {hundredreward/100:.2f}")
            hundredreward=0.0
        state,_=env.reset()
        st=torch.tensor(state,dtype=torch.float32,device=device).unsqueeze(0)
    if step>=learning_starts and step%train_freq==0:
        optimize() 
    if step%target_update==0:
        target_net.load_state_dict(policy_net.state_dict())

torch.save(policy_net.state_dict(),"dqn_mountaincar_model.pt")
```

```python
from PIL import Image
from IPython import display as ipythondisplay

env_render = MoutainCarRewardWrapper(gym.make("MountainCar-v0", render_mode='rgb_array'))
state, _ = env_render.reset()
done = False
total_reward = 0

policy_net.eval()

print("Starting Visual Evaluation...")

while not done:
    frame = env_render.render()
    img = Image.fromarray(frame)
    
    ipythondisplay.clear_output(wait=True)
    ipythondisplay.display(img)

    st = torch.tensor(state, dtype=torch.float32, device=device).unsqueeze(0)
    
    with torch.no_grad():
        action = select_action(st, total_steps + 1) 
        
    obs, reward, terminated, truncated, _ = env_render.step(action)
    
    total_reward += reward
    state = obs
    done = terminated or truncated

print(f"Finished! Evaluation Reward: {total_reward:.2f}")
env_render.close()
```
